## Histogrammes empil√©s

Les histogrammes empil√©s (*stacked histograms*) consistent √† empiler chaque colonne l'une sur l'autre.

```python
# on peut superposer des histogrammes, mais ce n'est pas exactement ce qu'on veut...
plt.hist(movies[movies.Genre=='Action'].Budget, bins=12)
plt.hist(movies[movies.Genre=='Drama'].Budget, bins=12)
plt.hist(movies[movies.Genre=='Thriller'].Budget, bins=12)
plt.show()
```

```python
# on peut mettre c√¥te √† c√¥te sur un m√™me graphique des histogrammes, mais ce n'est toujours pas ce qu'on veut...
plt.hist([movies[movies.Genre=='Action'].Budget,
          movies[movies.Genre=='Drama'].Budget,
          movies[movies.Genre=='Thriller'].Budget],
         bins=12)
plt.show()
```

```python
movies
movies[['Film','Genre']]
movies.Film
```

```python
# une solution : l'option stacked
plt.hist([movies[movies.Genre=='Action'].Budget,
          movies[movies.Genre=='Drama'].Budget,
          movies[movies.Genre=='Thriller'].Budget],
         bins=15,
        stacked = True)
plt.show()
```

## Facet Grid

```python
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = {'s':50, 'linewidth':0.5, 'edgecolor':'black'}
#kws = dict(s=50, linewidth=0.5, edgecolor='black')          
g = g.map(plt.scatter, 'CriticRatings', 'AudienceRatings', **kws)
```

```python
# on peut g√©n√©raliser √† tous les types de graphiques
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')
kws = dict(bins=12)
g = g.map(plt.hist, 'CriticRatings', **kws)
```

```python
# on peut ajouter les diagonales par exemple et fixer les √©chelles des axes
g = sns.FacetGrid(movies, row='Genre', col='Year', hue='Genre')

kws = {'s':50, 'linewidth':0.5, 'edgecolor':'black'} #kws = dict(s=50, linewidth=0.5, edgecolor=black')

g = g.map(plt.scatter, 'CriticRatings', 'AudienceRatings', **kws)

g.set(xlim=(0,100), ylim=(0,100))

for ax in g.axes.flat:
    ax.plot((0,100),(0,100), c='gray', ls='--')
    
g.add_legend()
```

## Une belle figure : un dashboard (tableau de bord)

```python
sns.set_style('white') #pour changer le fond, essayer avec 'darkgrid' par exemple pour voir la diff√©rence

f,axes = plt.subplots(2,2,figsize =(15,15))

# Axes 1
k1 = sns.kdeplot(data=movies, x="Budget", y="AudienceRatings", ax = axes[0,0])

# Axes 2
k2 = sns.kdeplot(data=movies, x="Budget", y="CriticRatings", ax = axes[0,1])

# Axes 3
v = sns.violinplot(data = movies, x ='Year', y='CriticRatings', ax = axes[1,0])

# Axes 4
k3 = sns.kdeplot(data=movies, x="CriticRatings", y="AudienceRatings", cmap='Reds', ax= axes[1,1])
k3 = sns.kdeplot(data=movies, x="CriticRatings", y="AudienceRatings",
                   fill = True, thresh=0.05, cmap='Reds', ax= axes[1,1])

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```

```python
sns.set_style('darkgrid', {'axes.facecolor' : 'black'})

f,axes = plt.subplots(2,2,figsize =(15,15))

# Axes 1
k1 = sns.kdeplot(data=movies, x="Budget", y="AudienceRatings", cmap='cool', ax = axes[0,0])
k1b = sns.kdeplot(data=movies, x="Budget", y="AudienceRatings",
                 fill = True, thresh=0.05, cmap='inferno', ax = axes[0,0])

# Axes 2
k2 = sns.kdeplot(data=movies, x="Budget", y="CriticRatings", cmap='Blues', ax = axes[0,1])
k2b = sns.kdeplot(data=movies, x="Budget", y="CriticRatings", 
                 fill = True, thresh=0.05, cmap='Greens', ax = axes[0,1])

# Axes 3
v = sns.violinplot(data = movies, x ='Year', y='CriticRatings', hue='Year', ax = axes[1,0],
                  palette = 'YlOrRd', legend=False)

# Axes 4
k3 = sns.kdeplot(data=movies, x="CriticRatings", y="AudienceRatings", cmap='gist_gray_r', ax= axes[1,1])
k3b = sns.kdeplot(data=movies, x="CriticRatings", y="AudienceRatings",
                   fill = True, thresh=0.05, cmap='Blues', ax= axes[1,1])

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))
plt.show()
```
