```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
path = os.path.join('data', 'exercice_dataset.csv')
df = pd.read_csv(path)
```

```python
df.head()
```

```python
df.info()
```

```python
df_norm = df[['Genre','Gross % US', 'Studio' ]]
df_norm.head()
```

```python
genres = ['action', 'comedy', 'adventure', 'animation', 'drama']
filtre_genres = df.Genre.isin(genres)
df1 = df[filtre_genres]
```

```python
studios = ['Buena Vista Studio', 'Sony', 'Universal', 'WB', 'Paramount Pictures', 'Fox']
filtre_studios = df1.Studio.isin(studios)
df2 = df1[filtre_studios]
```

https://www.statology.org/seaborn-legend-position/#:~:text=To%20change%20the%20position%20of%20a%20legend%20in%20a%20seaborn,legend()%20command.&text=The%20default%20location%20is%20%E2%80%9Cbest,avoids%20covering%20any%20data%20points.

```python
import numpy as np
fig= plt.figure(figsize=(15, 10))
vs1 = sns.boxplot(data=df1, x='Genre', y='Gross % US')
vs2 = sns.stripplot(data=df2,x='Genre', y='Gross % US', hue='Studio', size=10)
plt.title("Domestic Gross % by Genre", size=30)
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.ylabel("Gross % US", fontsize=20)
plt.xlabel("Genre", fontsize = 20)
plt.show()
```
